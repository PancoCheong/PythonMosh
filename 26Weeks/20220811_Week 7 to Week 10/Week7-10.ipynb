## 9.0.2 Pivot_Table

Create a spreadsheet-style pivot table (supports data aggregation)

```python
# Aggregates values by taking the sum
table = pd.pivot_table (df, values = 'RCD', index=['Property','Segment'],columns='GameType',aggfunc=np.sum)
table
```

```python
# Fill missing values using the fill_value parameter
table2 = pd.pivot_table(df, values='RCD', index=['Property','Segment'],
                    columns=['GameType'], aggfunc=np.sum, fill_value=0)

table2
```

```python
# Make a pivot table and add an aggregate function 
df.pivot_table(index='Property',aggfunc="sum")
```

```python
# Make a pivot table and add an aggregate function and select some columns
df.pivot_table(index='Property',values=['RCD','TimePlayed'],aggfunc="sum")
```

```python
# Aggregates by taking the mean across multiple columns
table_mean = pd.pivot_table(df, values=['RCD', 'TimePlayed'], index=['Property', 'Segment'],
                    aggfunc={'RCD': np.mean,
                             'TimePlayed': np.mean})

table_mean
```

```python
# Calculate multiple types of aggregations for any given value column
table3 = pd.pivot_table(df, values=['RCD', 'TimePlayed'], index=['Property', 'Segment'],
                    aggfunc={'RCD': np.mean,
                             'TimePlayed': [min, max, np.mean]})
table3
```

```python
df.groupby('Sales Rep').agg({  

'Order Id':'size', 

'Val':['sum','mean'], 

'Sale':['sum','mean'] 

}) 

#'size': Counts the rows 
#'sum': Sums the column up 
#'mean'/'median': Mean/Median of the column 
#'max'/'min': Maximum/Minimum of the column 
#'idxmax'/'idxmin': Index of the maximum/minimum of the column. Getting the index of the minimal or maximal value is helpful for mapping other columns, i.e., what's the name of the company per sales rep with the biggest deal 
# pd.Series.nunique: Counts unique values. Note that, unlike the previous functions, this is an actual function and not a string. 
```

```python
def cr(x): 

 return round(np.mean(x),2) 

# Long Form: Explictly specifying the NamedAgg 
aggregation = { 
'Potential Sales': pd.NamedAgg(column='Val', aggfunc='size'), 
'Sales': pd.NamedAgg(column='Sale', aggfunc='sum'), 
'Conversion Rate': pd.NamedAgg(column='Sale', aggfunc=cr) 
} 

# Alternative: Since the NamedAgg is just a tuple, we can also pass regular tuples 
aggregation = { 
'Potential Sales': ('Val','size'), 
'Sales': ('Sale','sum'), 
'Conversion Rate': ('Sale',cr) 
} 


df.groupby('Sales Rep').agg(**aggregation) 
```

```python
df.groupby('Sales Rep')['Val'].transform(lambda x: x/sum(x)) 
#While agg returns a reduced version of the input, transform returns an on a group-level 
# transformed version of the full data. The new output data has the same length as the input data.
```

```python
df['%'] = df.groupby('Sales Rep')['Val'].transform( 

lambda x: x/sum(x) 

) 
```

```python
df.groupby('Sales Rep').filter( 

lambda x: (x['Val'] * x['Sale']).sum() > 200000 

) 
```
